SUBESPACIOS VECTORIALES DE UNA TRANSFORMACIÓN LINEAL


T: ℝ³ -> ℝ³ está determinado por T(x, y, z) = (x - y, 2x + y + z, 3y - z). 
  1. Determinar si T es una transformación lineal.  2. Encuentra el núcleo y la imagen de T.  3. Compruebe si el kernel y la imagen son subespacios vectoriales.  
 Solución: 
 1. Para determinar si T es una transformación lineal, se deben verificar dos propiedades: la  suma conservativa y la  multiplicación que conserva el escalar. 
  - Garantía total: 
 T((x₁, y₁, z₁) + (x₂, y₂, z₂)) = T(x₁ + x₂, y₁ + y₂, z₁ + z₂) 
 = ((x₁ + x₂) - (y₁ + y₂), 2(x₁ + x₂) + (y₁ + y₂) + (z₁ + z₂), 3(y₁ + y₂) - (z₁ + z₂)) 
 = ((x₁ - y₁) + (x₂ - y₂), 2x₁ + y₁ + z₁ + 2x₂ + y₂ + z₂, 3y₁ - z₁ + 3y₂ - z₂) 
 = (x₁ - y₁, 2x₁ + y₁ + z₁, 3y₁ - z₁) + (x₂ - y₂, 2x₂ + y₂ + z₂, 3y₂ - z₂) 
 = T(x₁, y₁, z₁) + T(x₂, y₂, z₂) 
 
 - Conservar la multiplicación con una cantidad: 
 T(c(x, y, z)) = T(cx, cy, cz) = (cx - cy, 2cx + cy + cz, 3cy - cz) 
 = c(x - y, 2x + y + z, 3y - z) 
 = cT(x, y, z) 
 
 Como T satisface ambas propiedades, podemos concluir que T es una transformación lineal.  2. Para encontrar la multiplicación de T, necesitamos encontrar todos los vectores (x, y, z) en ℝ³ tales que T(x, y, z) = (0, 0, 0). 
  Resolviendo las ecuaciones, obtenemos: 
 x-y=0 
 2x + y + z = 0 
 3y - z = 0 
 
 De la primera ecuación, tenemos  x = y. Sustituyendo esto en la segunda y tercera ecuaciones obtenemos: 
 2y + y + z = 0 
 3y - z = 0 
 
 Resolviendo estas dos ecuaciones simultáneas, obtenemos: 
 y = z/3 
 x = y = z/3 
 
 Por  tanto, la multiplicación de T viene dada por todos los vectores de la forma (z/3, z/3, z) para todos los valores de z.  
 La imagen de T  dada por todos los vectores en ℝ³ tiene la forma T(x, y, z) = (x - y, 2)

```python
import numpy as np

def transformation(x, y, z):
    transformed_x = x - y
    transformed_y = 2*x + y + z
    transformed_z = 3*y - z
    return transformed_x, transformed_y, transformed_z

# Verificar si la transformación es lineal
def is_linear_transformation():
    x1, y1, z1 = np.random.rand(3)  # Vector 1
    x2, y2, z2 = np.random.rand(3)  # Vector 2
    c = np.random.rand()  # Escalar aleatorio

    v1 = transformation(x1, y1, z1)
    v2 = transformation(x2, y2, z2)
    v_sum = transformation(x1 + x2, y1 + y2, z1 + z2)
    v_scalar = transformation(c*x1, c*y1, c*z1)

    if v_sum == tuple(np.array(v1) + np.array(v2)) and v_scalar == tuple(c*np.array(v1)):
        return True
    else:
        return False

# Encontrar el núcleo de la transformación
def find_null_space():
    null_space = []
    for z in np.arange(-10, 10):
        x = y = z/3
        v = transformation(x, y, z)
        if np.all(np.array(v) == np.array([0, 0, 0])):
            null_space.append((x, y, z))
    return null_space

# Encontrar la imagen de la transformación
def find_image():
    image = []
    for x in np.arange(-10, 10):
        for y in np.arange(-10, 10):
            for z in np.arange(-10, 10):
                v = transformation(x, y, z)
                if np.all(np.array(v) != np.array([0, 0, 0])):
                    image.append((x, y, z))
    return image

# Verificar si el núcleo y la imagen son subespacios vectoriales
def is_subspace(vector_list):
    zero_vector = (0, 0, 0)
    if zero_vector in vector_list:
        vector_list.remove(zero_vector)
    for i in range(len(vector_list)):
        for j in range(i+1, len(vector_list)):
            if np.all(np.array(vector_list[i]) + np.array(vector_list[j]) != np.array(zero_vector)):
                return False
    return True

# Verificación de la transformación lineal
is_linear = is_linear_transformation()
print("¿Es la transformación una transformación lineal? ", is_linear)

# Encontrar el núcleo de la transformación
null_space = find_null_space()
print("Núcleo de la transformación: ", null_space)

# Encontrar la imagen de la transformación
image = find_image()
print("Imagen de la transformación: ", image)

# Verificar si el núcleo y la imagen son subespacios vectoriales
is_null_space_subspace = is_subspace(null_space)
is_image_subspace = is_subspace(image)
print("¿El núcleo es un subespacio vectorial? ", is_null_space_subspace)
print("¿La imagen es un subespacio vectorial? ", is_image_subspace)
```

TRANSFORMACIONES LINEALES

 T: R^2 -> R^2 definida por la siguiente matriz:

A = [1 2]
    [3 4]

1. Encuentra la imagen del vector (2, 1) bajo la transformación T.
2. Encuentra la matriz de la transformación lineal T^2 (la composición de T consigo misma).
3. Encuentra la matriz de la transformación lineal T^-1 (la inversa de T) si existe.

Solución:
1. Para encontrar la imagen del vector (2, 1) bajo la transformación T, multiplicamos la matriz A por el vector columna (2, 1):

A * [2] = [1*2 + 2*1] = [4]
      [1*3 + 2*4]   [11]

Por lo tanto, la imagen de (2, 1) bajo la transformación T es el vector (4, 11).

2. Para encontrar la matriz de la transformación lineal T^2, simplemente multiplicamos la matriz A consigo misma:

A^2 = A * A = [1 2] * [1 2] = [1*1 + 2*3 1*2 + 2*4] = [7 10]
                [3 4]   [3*1 + 4*3 3*2 + 4*4]   [15 22]

Por lo tanto, la matriz de la transformación lineal T^2 es:

[7 10]
[15 22]

3. Para encontrar la matriz de la transformación lineal T^-1 (inversa de T), primero verifiquemos si T es invertible. Para ello, calculamos el determinante de la matriz A:

det(A) = (1*4) - (2*3) = 4 - 6 = -2

El determinante es diferente de cero (-2 ≠ 0), por lo que la matriz A es invertible. Para encontrar la inversa de A, podemos utilizar la fórmula:

A^-1 = (1/det(A)) * adj(A)

Donde adj(A) es la matriz adjunta de A. Calculamos la matriz adjunta de A intercambiando los elementos diagonales, cambiando el signo de los elementos no diagonales y transponiendo la matriz resultante:

adj(A) = [4 -2]
         [-3 1]

Finalmente, calculamos la inversa de A:

A^-1 = (1/(-2)) * [4 -2] = [-2 1]
                   [-3 1]   [3/2 -1/2]

Por lo tanto, la matriz de la transformación lineal T^-1 es:

[-2 1]
[3/2 -1/2]

VECTOR PROPIO

En álgebra lineal, un vector propio (también conocido como autovector) es un vector no nulo que, cuando se multiplica por una matriz, da lugar a un múltiplo escalar del mismo vector. Es decir, si A es una matriz cuadrada y v es un vector no nulo, entonces v es un vector propio de A si existe un número escalar λ tal que:

Av = λv

Aquí, Av representa el producto matricial de A y v, y λ es el valor propio correspondiente al vector propio v.

La importancia de los vectores propios radica en que ofrecen información valiosa sobre las transformaciones lineales representadas por las matrices. Los vectores propios representan direcciones especiales a lo largo de las cuales una transformación lineal solo causa un estiramiento o compresión, sin cambiar la dirección. El escalar λ, llamado valor propio o autovalor, representa el factor por el cual el vector propio se estira o comprime.

Al calcular los vectores propios y los valores propios de una matriz, se pueden obtener características fundamentales de la matriz, como su diagonalización, simetría, determinantes y trazas. Además, los vectores propios se utilizan en una variedad de aplicaciones prácticas, como en el análisis de sistemas dinámicos, la resolución de ecuaciones diferenciales y la compresión de datos, entre otros.

```python
import numpy as np

# Definir la matriz
A = np.array([[2, -1], [4, 3]])

# Calcular los vectores propios y los valores propios
eigenvalues, eigenvectors = np.linalg.eig(A)

# Imprimir los valores propios
print("Valores propios:")
for eigenvalue in eigenvalues:
    print(eigenvalue)

# Imprimir los vectores propios
print("\nVectores propios:")
for i in range(len(eigenvectors)):
    eigenvector = eigenvectors[:, i]
    print(eigenvector)
```
